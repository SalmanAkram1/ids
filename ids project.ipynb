```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Load the dataset
df=pd.read_csv('./sampled_dataset.csv')

# Display the number of missing values in each column
print(df.isnull().sum())

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Plot boxplots for numerical features to detect outliers
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

# Address outliers by capping them
def cap_outliers(series, lower_quantile=0.01, upper_quantile=0.99):
    lower_limit = series.quantile(lower_quantile)
    upper_limit = series.quantile(upper_quantile)
    return np.clip(series, lower_limit, upper_limit)

df[numerical_columns] = df[numerical_columns].apply(cap_outliers)
```

```python
# Fill object columns with mode
object_columns = ['DataValueFootnoteSymbol', 'DataValueFootnote', 'Geolocation']
for col in object_columns:
    mode_value = df[col].mode()[0]
    df[col] = df[col].fillna(mode_value)
```

```python
# Fill float columns with mean
float_columns = ['DataValue', 'DataValueAlt', 'LowConfidenceLimit', 'HighConfidenceLimit']
for col in float_columns:
    mean_value = df[col].mean()
    df[col] = df[col].fillna(mean_value)
```

```python
features = ['DataValue', 'HighConfidenceLimit', 'DataValueAlt', 'LowConfidenceLimit']

# Initialize dictionaries to store quartile values
Q1 = {}
Q3 = {}
IQR = {}

# Iterate over each column in the features list
for column in features:
    # Q1 and Q3 calculations, and the interquartile range (IQR) for the current column
    Q1[column] = df[column].quantile(0.25)
    Q3[column] = df[column].quantile(0.75)
    IQR[column] = Q3[column] - Q1[column]
    print("Interquartile Range (IQR) for", column, ":", IQR[column])
```

```python
for column in features:
    df = df[~((df[column] < (Q1[column] - 1.5 * IQR[column])) | (df[column] > (Q3[column] + 1.5 * IQR[column])))]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Select features and target
features = ['YearStart', 'YearEnd', 'LowConfidenceLimit', 'HighConfidenceLimit']
target = 'DataValue'

# Remove rows with missing values in selected columns
df_ml = df.dropna(subset=features + [target])

X = df_ml[features]
y = df_ml[target]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Model 1: RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_train_pred_rf = rf_model.predict(X_train)
y_test_pred_rf = rf_model.predict(X_test)

# Evaluate RandomForestRegressor
mse_train_rf = mean_squared_error(y_train, y_train_pred_rf)
r2_train_rf = r2_score(y_train, y_train_pred_rf)

mse_test_rf = mean_squared_error(y_test, y_test_pred_rf)
r2_test_rf = r2_score(y_test, y_test_pred_rf)

print(f"Random Forest Regressor Training MSE: {mse_train_rf}")
print(f"Random Forest Regressor Training RÂ²: {r2_train_rf}")
print(f"Random Forest Regressor Testing MSE: {mse_test_rf}")
print(f"Random Forest Regressor Testing RÂ²: {r2_test_rf}")
```

```python
# Model 2: LinearRegression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predictions
y_train_pred_lr = lr_model.predict(X_train)
y_test_pred_lr = lr_model.predict(X_test)

# Evaluate LinearRegression
mse_train_lr = mean_squared_error(y_train, y_train_pred_lr)
r2_train_lr = r2_score(y_train, y_train_pred_lr)

mse_test_lr = mean_squared_error(y_test, y_test_pred_lr)
r2_test_lr = r2_score(y_test, y_test_pred_lr)

print(f"Linear Regression Training MSE: {mse_train_lr}")
print(f"Linear Regression Training RÂ²: {r2_train_lr}")
print(f"Linear Regression Testing MSE: {mse_test_lr}")
print(f"Linear Regression Testing RÂ²: {r2_test_lr}")
```
